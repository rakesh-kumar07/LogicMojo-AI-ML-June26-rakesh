## 1. INT (Integer) Data Type

Integers are whole numbers. Let's explore arithmetic operations, type checking, and bitwise operations.

```python
# Q1: Basic arithmetic operations
a = 15
b = 4

print(f"a = {a}, b = {b}")
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponent: {a} ** {b} = {a ** b}")
```

```python
# Q2: Integer type checking and conversion
x = 42
print(f"Value: {x}, Type: {type(x)}")
print(f"Is integer: {isinstance(x, int)}")
print(f"Absolute value of -50: {abs(-50)}")
```

```python
# Q3: Bitwise operations
num1 = 12  # Binary: 1100
num2 = 10  # Binary: 1010

print(f"num1 = {num1} (binary: {bin(num1)})")
print(f"num2 = {num2} (binary: {bin(num2)})")
print(f"Bitwise AND: {num1 & num2} (binary: {bin(num1 & num2)})")
print(f"Bitwise OR: {num1 | num2} (binary: {bin(num1 | num2)})")
print(f"Bitwise XOR: {num1 ^ num2} (binary: {bin(num1 ^ num2)})")
```

## 2. FLOAT (Floating-Point Numbers)

Floats represent decimal numbers. Let's practice mathematical operations, precision, and conversions.

```python
# Q1: Basic float operations
pi = 3.14159
radius = 5.0

print(f"Pi: {pi}, Radius: {radius}")
print(f"Circumference: 2 * π * r = {2 * pi * radius}")
print(f"Area of circle: π * r² = {pi * (radius ** 2)}")
```

```python
# Q2: Float precision and rounding
value = 10.23456789
print(f"Original value: {value}")
print(f"Rounded to 2 decimals: {round(value, 2)}")
print(f"Rounded to 4 decimals: {round(value, 4)}")
```

## 3. LIST (Ordered, Mutable Sequences)

Lists allow duplicates and can be modified. Let's explore indexing, methods, slicing, and comprehensions.

```python
# Q1: Creating lists and accessing elements
fruits = ["apple", "banana", "cherry", "date", "apple"]

print(f"List: {fruits}")
print(f"First element (index 0): {fruits[0]}")
print(f"Last element (index -1): {fruits[-1]}")
print(f"Length: {len(fruits)}")
```

```python
# Q2: Modifying lists - append, extend, insert, remove, pop
numbers = [1, 2, 3]
print(f"Original: {numbers}")

numbers.append(4)
print(f"After append(4): {numbers}")

numbers.extend([5, 6])
print(f"After extend([5, 6]): {numbers}")

numbers.insert(0, 0)
print(f"After insert(0, 0): {numbers}")

numbers.remove(3)
print(f"After remove(3): {numbers}")

popped = numbers.pop()
print(f"After pop(): {numbers}, Popped value: {popped}")
```

```python
# Q3: List slicing operations
letters = ['a', 'b', 'c', 'd', 'e', 'f']

print(f"Original list: {letters}")
print(f"letters[1:4]: {letters[1:4]}")
print(f"letters[::2]: {letters[::2]}")
print(f"letters[::-1]: {letters[::-1]}")
```

```python
# Q4: List methods - count, index, sort, reverse
nums = [3, 1, 4, 1, 5, 9, 2, 6]

print(f"Original: {nums}")
print(f"Count of 1: {nums.count(1)}")
print(f"Index of 5: {nums.index(5)}")

sorted_nums = sorted(nums)
print(f"Sorted: {sorted_nums}")

nums.reverse()
print(f"After reverse(): {nums}")
```

```python
# Q5: List comprehension
squares = [x**2 for x in range(1, 6)]
print(f"Squares 1-5: {squares}")

evens = [x for x in range(10) if x % 2 == 0]
print(f"Even numbers 0-9: {evens}")
```

```python
# Q6: More list operations - clear, copy, in operator, min, max, sum
numbers = [5, 2, 8, 1, 9, 3]
print(f"Original list: {numbers}")

# Using min, max, sum
print(f"Minimum value: {min(numbers)}")
print(f"Maximum value: {max(numbers)}")
print(f"Sum of all elements: {sum(numbers)}")

# Membership testing with 'in'
print(f"Is 8 in list: {8 in numbers}")
print(f"Is 10 in list: {10 in numbers}")

# Copy a list
numbers_copy = numbers.copy()
print(f"Copy of list: {numbers_copy}")

# Modify copy without affecting original
numbers_copy[0] = 999
print(f"After modifying copy - Original: {numbers}, Copy: {numbers_copy}")

# Clear a list
temp_list = [1, 2, 3, 4]
print(f"Before clear: {temp_list}")
temp_list.clear()
print(f"After clear: {temp_list}")
```

```python
# Q7: Advanced list operations - enumerate, zip, list concatenation, repetition
# Enumerate - get index and value
fruits = ["apple", "banana", "cherry"]
print("Using enumerate:")
for index, fruit in enumerate(fruits):
    print(f"  Index {index}: {fruit}")

# Zip - combine multiple lists
names = ["Alice", "Bob", "Charlie"]
scores = [95, 87, 92]
print("\nUsing zip - combining lists:")
for name, score in zip(names, scores):
    print(f"  {name}: {score}")

# List concatenation with +
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2
print(f"\nList concatenation: {list1} + {list2} = {combined}")

# List repetition with *
repeated = [1, 2] * 3
print(f"List repetition: [1, 2] * 3 = {repeated}")

# Join lists without creating new list (extend)
list_a = [10, 20]
list_b = [30, 40]
list_a.extend(list_b)
print(f"After extend: {list_a}")

# Changing elements by index and slicing
items = ['a', 'b', 'c', 'd', 'e']
print(f"Original: {items}")
items[1] = 'B'
print(f"After changing index 1: {items}")
items[1:3] = ['B', 'C']
print(f"After changing slice [1:3]: {items}")
```

```python
# Q8: Difference between remove(), pop(), and del for removing elements
print("=== REMOVING ELEMENTS: remove() vs pop() vs del ===")

# remove() - removes by VALUE
numbers1 = [1, 2, 3, 4, 5, 3]
print(f"\nOriginal list: {numbers1}")
numbers1.remove(3)  # Removes FIRST occurrence of value 3
print(f"After remove(3): {numbers1}")
print("Note: remove() removes by VALUE, not index")

# pop() - removes by INDEX and returns the value
numbers2 = [1, 2, 3, 4, 5]
print(f"\nOriginal list: {numbers2}")
removed_val = numbers2.pop(2)  # Removes element at index 2 (value 3)
print(f"After pop(2): {numbers2}")
print(f"Returned value: {removed_val}")
print("Note: pop() removes by INDEX and returns the removed value")

# del - removes by INDEX (no return value)
numbers3 = [1, 2, 3, 4, 5]
print(f"\nOriginal list: {numbers3}")
del numbers3[2]  # Removes element at index 2 (value 3)
print(f"After del numbers3[2]: {numbers3}")
print("Note: del removes by INDEX, doesn't return anything")

# Comparison table
print("\n" + "="*60)
print("COMPARISON TABLE:")
print("="*60)
print(f"{'Method':<15} {'Removes by':<15} {'Returns Value':<15}")
print("-"*60)
print(f"{'remove(value)':<15} {'VALUE':<15} {'No':<15}")
print(f"{'pop(index)':<15} {'INDEX':<15} {'Yes':<15}")
print(f"{'del list[i]':<15} {'INDEX':<15} {'No':<15}")
print("="*60)
```

## 4. TUPLE (Ordered, Immutable Sequences)

Tuples are immutable (cannot be changed) but allow duplicates. Let's explore unpacking and immutability.

```python
# Q1: Creating tuples and accessing elements
coordinates = (10, 20, 30)
student = ("Alice", 25, 3.8, True)

print(f"Coordinates: {coordinates}")
print(f"Student info: {student}")
print(f"First element: {coordinates[0]}")
print(f"Last element: {coordinates[-1]}")
print(f"Length: {len(coordinates)}")
```

```python
# Q2: Tuple unpacking
point = (5, 15)
x, y = point
print(f"Point: {point}")
print(f"Unpacked - x: {x}, y: {y}")

a, b, c = (100, 200, 300)
print(f"Unpacked - a: {a}, b: {b}, c: {c}")
```

```python
# Q3: Tuple immutability
my_tuple = (1, 2, 3)
print(f"Original tuple: {my_tuple}")

try:
    my_tuple[0] = 10
except TypeError as e:
    print(f"Error when trying to modify: {e}")
```

```python
# Q4: Tuple methods - count and index
colors = ("red", "blue", "red", "green", "blue", "red")

print(f"Tuple: {colors}")
print(f"Count of 'red': {colors.count('red')}")
print(f"Index of 'green': {colors.index('green')}")
```

```python
# Q5: Single element tuple (with trailing comma)
single = (42,)  # Note the comma - this is a tuple!
print(f"Single element tuple: {single}, Type: {type(single)}")

not_tuple = (42)  # This is just an integer in parentheses
print(f"Without comma: {not_tuple}, Type: {type(not_tuple)}")

single_element_tuples = [
    (1,),
    (2,),
    (3,)
]
print(f"List of single-element tuples: {single_element_tuples}")
```

```python
# Q6: Nested tuples and tuple operations
# Nested tuples
nested = ((1, 2), (3, 4), (5, 6))
print(f"Nested tuple: {nested}")
print(f"First nested tuple: {nested[0]}")
print(f"Element at [0][1]: {nested[0][1]}")

# Tuple concatenation with +
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined_tuple = tuple1 + tuple2
print(f"\nTuple concatenation: {tuple1} + {tuple2} = {combined_tuple}")

# Tuple repetition with *
repeated_tuple = (1, 2) * 3
print(f"Tuple repetition: (1, 2) * 3 = {repeated_tuple}")

# Slicing tuples
data = (10, 20, 30, 40, 50)
print(f"\nOriginal tuple: {data}")
print(f"data[1:4]: {data[1:4]}")
print(f"data[::2]: {data[::2]}")
print(f"data[::-1]: {data[::-1]}")

# Membership testing with 'in'
print(f"\n30 in data: {30 in data}")
print(f"100 in data: {100 in data}")
```

```python
# Q7: Converting between tuple and list
# Tuple to List
original_tuple = (1, 2, 3, 4, 5)
converted_list = list(original_tuple)
print(f"Original tuple: {original_tuple}")
print(f"Converted to list: {converted_list}")

# Modify the list
converted_list[0] = 10
converted_list.append(6)
print(f"Modified list: {converted_list}")

# Convert back to tuple
new_tuple = tuple(converted_list)
print(f"Converted back to tuple: {new_tuple}")

# Finding min, max, sum in tuples
numbers_tuple = (45, 23, 98, 12, 56)
print(f"\nTuple of numbers: {numbers_tuple}")
print(f"Minimum: {min(numbers_tuple)}")
print(f"Maximum: {max(numbers_tuple)}")
print(f"Sum: {sum(numbers_tuple)}")
print(f"Length: {len(numbers_tuple)}")

# Enumerating through tuples
colors = ("red", "green", "blue")
print("\nEnumerating through tuple:")
for index, color in enumerate(colors):
    print(f"  {index}: {color}")
```

## 5. DICTIONARY (Key-Value Pairs)

Dictionaries store data as key-value pairs. Let's explore creation, access, modification, and iteration.

```python
# Q1: Creating dictionaries and accessing values
person = {"name": "John", "age": 30, "city": "New York", "age": 31}

print(f"Dictionary: {person}")
print(f"Name: {person['name']}")
print(f"Age: {person['age']}")
print(f"Get city with default: {person.get('country', 'Unknown')}")
```

```python
# Q2: Adding and modifying dictionary items
student_info = {"name": "Bob", "grade": "A"}
print(f"Original: {student_info}")

student_info["age"] = 20
print(f"After adding age: {student_info}")

student_info["grade"] = "A+"
print(f"After modifying grade: {student_info}")
```

```python
# Q3: Removing dictionary items
data = {"a": 1, "b": 2, "c": 3, "d": 4}
print(f"Original: {data}")

del data["b"]
print(f"After del data['b']: {data}")

popped_value = data.pop("c")
print(f"After pop('c'): {data}, Popped: {popped_value}")
```

```python
# Q4: Dictionary methods - keys, values, items
scores = {"Alice": 95, "Bob": 87, "Charlie": 92}

print(f"Dictionary: {scores}")
print(f"Keys: {list(scores.keys())}")
print(f"Values: {list(scores.values())}")
print(f"Items: {list(scores.items())}")
```

```python
# Q5: Dictionary iteration
scores = {"Alice": 95, "Bob": 87, "Charlie": 92}

for key in scores:
    print(f"  {key}: {scores[key]}")
```

```python
# Q6: Dictionary comprehension
numbers_dict = {x: x**2 for x in range(1, 6)}
print(f"Squares dict: {numbers_dict}")
```

## 6. SET (Unordered, Mutable, No Duplicates)

Sets contain unique elements and support mathematical set operations.

```python
# Q1: Creating sets and removing duplicates
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unique_set = set(numbers)

print(f"List with duplicates: {numbers}")
print(f"Set (duplicates removed): {unique_set}")

colors_set = {"red", "blue", "green", "red"}
print(f"Colors set: {colors_set}")
```

```python
# Q2: Adding and removing elements in sets
fruits_set = {"apple", "banana"}
print(f"Original: {fruits_set}")

fruits_set.add("cherry")
print(f"After add('cherry'): {fruits_set}")

fruits_set.discard("banana")
print(f"After discard('banana'): {fruits_set}")
```

```python
# Q3: Set operations - union, intersection, difference
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

print(f"Set A: {set_a}")
print(f"Set B: {set_b}")
print(f"Union (A | B): {set_a | set_b}")
print(f"Intersection (A & B): {set_a & set_b}")
print(f"Difference (A - B): {set_a - set_b}")
print(f"Symmetric Difference (A ^ B): {set_a ^ set_b}")
```

```python
# Q4: Set membership and subset/superset
s1 = {1, 2, 3}
s2 = {1, 2}

print(f"Set 1: {s1}")
print(f"Set 2: {s2}")
print(f"2 in Set 1: {2 in s1}")
print(f"Is Set 2 subset of Set 1: {s2.issubset(s1)}")
print(f"Is Set 1 superset of Set 2: {s1.issuperset(s2)}")
```

```python
# Q5: Set comprehension
even_set = {x for x in range(10) if x % 2 == 0}
print(f"Even numbers 0-9 (set): {even_set}")
```

## 7. FROZENSET (Unordered, Immutable, No Duplicates)

Frozensets are like sets but immutable. They can be used as dictionary keys.

```python
# Q1: Creating frozensets
list_data = [1, 2, 3, 2, 1]
fs = frozenset(list_data)

print(f"List: {list_data}")
print(f"Frozenset: {fs}")

colors_fs = frozenset(["red", "blue", "green"])
print(f"Colors frozenset: {colors_fs}")
```

```python
# Q2: Frozenset immutability
my_frozenset = frozenset({1, 2, 3})
print(f"Frozenset: {my_frozenset}")

try:
    my_frozenset.add(4)
except AttributeError as e:
    print(f"Cannot add to frozenset: frozenset has no 'add' method")
```

```python
# Q3: Frozenset operations - union, intersection, difference
fs_a = frozenset({1, 2, 3, 4})
fs_b = frozenset({3, 4, 5, 6})

print(f"Frozenset A: {fs_a}")
print(f"Frozenset B: {fs_b}")
print(f"Union: {fs_a | fs_b}")
print(f"Intersection: {fs_a & fs_b}")
print(f"Difference: {fs_a - fs_b}")
```

```python
# Q4: Frozenset as dictionary key
coordinates_dict = {
    frozenset({1, 2}): "Point A",
    frozenset({3, 4}): "Point B"
}
print(f"Dictionary with frozenset keys: {coordinates_dict}")

# Try with regular set (will fail)
try:
    regular_set_key = {set({1, 2}): "value"}
except TypeError as e:
    print(f"Cannot use set as dict key (unhashable): {e}")
```

```python
# Q5: Frozenset membership and operations
fs = frozenset([10, 20, 30])

print(f"Frozenset: {fs}")
print(f"20 in frozenset: {20 in fs}")
print(f"Length: {len(fs)}")
```

## 4. TUPLE (Ordered, Immutable Sequences)

Tuples are immutable (cannot be changed) but allow duplicates. Let's explore unpacking and immutability.

```python
# Q1: Creating tuples and accessing elements
coordinates = (10, 20, 30)
student = ("Alice", 25, 3.8, True)

print(f"Coordinates: {coordinates}")
print(f"Student info: {student}")
print(f"First element: {coordinates[0]}")
print(f"Last element: {coordinates[-1]}")
print(f"Length: {len(coordinates)}")
```

```python
# Q2: Tuple unpacking
point = (5, 15)
x, y = point
print(f"Point: {point}")
print(f"Unpacked - x: {x}, y: {y}")

a, b, c = (100, 200, 300)
print(f"Unpacked - a: {a}, b: {b}, c: {c}")
```

```python
# Q3: Tuple immutability
my_tuple = (1, 2, 3)
print(f"Original tuple: {my_tuple}")

try:
    my_tuple[0] = 10
except TypeError as e:
    print(f"Error when trying to modify: {e}")
```

```python
# Q4: Tuple methods - count and index
colors = ("red", "blue", "red", "green", "blue", "red")

print(f"Tuple: {colors}")
print(f"Count of 'red': {colors.count('red')}")
print(f"Index of 'green': {colors.index('green')}")
```

```python
# Q5: Single element tuple (with trailing comma)
single = (42,)  # Note the comma - this is a tuple!
print(f"Single element tuple: {single}, Type: {type(single)}")

not_tuple = (42)  # This is just an integer in parentheses
print(f"Without comma: {not_tuple}, Type: {type(not_tuple)}")

single_element_tuples = [
    (1,),
    (2,),
    (3,)
]
print(f"List of single-element tuples: {single_element_tuples}")
```

```python
# Q6: Nested tuples and tuple operations
# Nested tuples
nested = ((1, 2), (3, 4), (5, 6))
print(f"Nested tuple: {nested}")
print(f"First nested tuple: {nested[0]}")
print(f"Element at [0][1]: {nested[0][1]}")

# Tuple concatenation with +
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined_tuple = tuple1 + tuple2
print(f"\nTuple concatenation: {tuple1} + {tuple2} = {combined_tuple}")

# Tuple repetition with *
repeated_tuple = (1, 2) * 3
print(f"Tuple repetition: (1, 2) * 3 = {repeated_tuple}")

# Slicing tuples
data = (10, 20, 30, 40, 50)
print(f"\nOriginal tuple: {data}")
print(f"data[1:4]: {data[1:4]}")
print(f"data[::2]: {data[::2]}")
print(f"data[::-1]: {data[::-1]}")

# Membership testing with 'in'
print(f"\n30 in data: {30 in data}")
print(f"100 in data: {100 in data}")
```

```python
# Q7: Converting between tuple and list
# Tuple to List
original_tuple = (1, 2, 3, 4, 5)
converted_list = list(original_tuple)
print(f"Original tuple: {original_tuple}")
print(f"Converted to list: {converted_list}")

# Modify the list
converted_list[0] = 10
converted_list.append(6)
print(f"Modified list: {converted_list}")

# Convert back to tuple
new_tuple = tuple(converted_list)
print(f"Converted back to tuple: {new_tuple}")

# Finding min, max, sum in tuples
numbers_tuple = (45, 23, 98, 12, 56)
print(f"\nTuple of numbers: {numbers_tuple}")
print(f"Minimum: {min(numbers_tuple)}")
print(f"Maximum: {max(numbers_tuple)}")
print(f"Sum: {sum(numbers_tuple)}")
print(f"Length: {len(numbers_tuple)}")

# Enumerating through tuples
colors = ("red", "green", "blue")
print("\nEnumerating through tuple:")
for index, color in enumerate(colors):
    print(f"  {index}: {color}")
```
